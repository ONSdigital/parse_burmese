```python
def get_all_translations():
    """
    Grabs all of the official English from the downloaded "en" translations yaml files 
    which will provide:
    1) the English text 
    2) the corresponding key. 
    
    The keys can be used to build the Burmese yaml later. 
    
    returns:
        dict of all data from yaml files
    """
    source_yaml_dir = "official_translations\\en"
    all_yamls = {}
    for file in os.listdir(source_yaml_dir):
        file_name = file[:-4]
        f = open(os.path.join(source_yaml_dir, file), 'r',  encoding="utf8")
        all_yamls[file_name] = {}
        for key,value in yaml.load(f).items():
            all_yamls[file_name][key] = value
    return all_yamls

all_yamls = get_all_translations()
```

```python
# Now we have all the translations from the yamls, we can make them into a dataframe
# This dataframe is needed for searching the values (translations) and picking the correct key
# Require structure is 
# --------------------------------------
# ¦ yamlfilename ¦ key_word ¦ sentence ¦
# ¦ -----------------------------------
all_translations = (pd.DataFrame(data=[(file_name, key_word, sentence)
                                       for file_name in all_yamls.keys()
                                       for key_word,sentence in all_yamls[file_name].items()],
                                 columns=['yaml_filename', 'key_word', 'sentence']))
```

```python
# straight copying this code from https://stackoverflow.com/a/56315491

def fuzzy_merge(df_1, df_2, key1, key2, threshold=90, limit=2):
    """
    :param df_1: the left table to join
    :param df_2: the right table to join
    :param key1: key column of the left table
    :param key2: key column of the right table
    :param threshold: how close the matches should be to return a match, based on Levenshtein distance
    :param limit: the amount of matches that will get returned, these are sorted high to low
    :return: dataframe with boths keys and matches
    """
    s = df_2[key2].tolist()

    m = df_1[key1].apply(lambda x: process.extract(x, s, limit=limit))    
    df_1['matches'] = m

    m2 = df_1['matches'].apply(lambda x: ', '.join([i[0] for i in x if i[1] >= threshold]))
    df_1['matches'] = m2

    return df_1

fuzzymerged_df = fuzzy_merge(all_translations, en_burm_trans, key1="en", key2="en")
```

```python
def search_for_word_in_translations(search):
    """
    Supply a word or phrase to search for and the function will search for the whole word
    in the all_translations df. It searches for chopped up sections of the phrase if the whole word is not found
    If more than
    """
#     import ipdb; ipdb.set_trace()
    key_word = reverse_lookup(column='key_word', search=search, df=all_translations)
    yaml_filename = reverse_lookup(column='yaml_filename', search=search, df=all_translations)
    if key_word and yaml_filename: #checking that both have been returned
        return key_word, yaml_filename
    else:
        #try chopping it up and taking and searching for a sub-section
        for i in range(5):
            chopped_search = make_chopped_search_term(search_term=search, divisor=i+1)
            key_word = reverse_lookup(column='key_word', search=chopped_search, df=all_translations)
            yaml_filename = reverse_lookup(column='yaml_filename', search=chopped_search, df=all_translations)
            if key_word and yaml_filename:
                return key_word, yaml_filename
        print (f'{search} --- not found in the en-en translations file')
#         ipdb.set_trace()

def reverse_lookup(column, search, df):
    """
    Searches the 'sentence' column for the search term and returns the result. 
    """
    #check if any exact matches of the search phrase are found
    if df.sentence.str.match(re.escape(search)).any(): # re.escape is used here because getting unbalanced parenthesis error
        #now need to check if there is more than 1 match
        if df[df.sentence.str.match(search)].shape[0]>1:
            print ("Error more than 1 found")
            #multiple results have been found. So should capture the dataframe
            df = df[df.sentence.str.match(search)]
            #it would be really good to insert the Burmese here : # df['burm'] = en_burm_trans[en_burm_trans.en.str.match(search)].burm
            print (df)
            return False 
        try:
            # using try because .item() throws a value error if the key doesn't exist
            key_word = df[df.sentence.str.match(search)].loc[:,f'{column}'].item() 
        except ValueError:  
            return "No translation found"
        return key_word
                                                    

def make_chopped_search_term(search_term, divisor=3):
    """
    Takes a search term (phrase, sentence) and divides it which should give greater chance of finding
    a partial match.
    """
    length = len(search_term)
    third_length = round(length/divisor)
    end = length+1
    highest_start = end - third_length
    random_start_point = random.randrange(0, highest_start)
    chopped_search = search_term[random_start_point:random_start_point+third_length]
    return chopped_search
```

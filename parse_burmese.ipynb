```python
import pandas as pd
import os
import yaml
import random
import re
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import numpy as np
```

```python
# importing the whole dataset, containing both English and Burmese translations
en_burm_trans = pd.DataFrame.from_csv("translations_for_myanmar.csv", header=None)
```

```python
# Seperating the English out (every other line starting at 0)
en = en_burm_trans.iloc[0::2]
```

```python
# Seperating the Burmese out (every other line starting at 1)
burm = en_burm_trans.iloc[1::2]
```

```python
# Checking it's worked. This will check if what's contained in the 'en' dataset are all A-Z western alphabet characters. If True, they are.
pattern = re.compile("[A-Za-z]+")
en.reset_index().loc[:,0].str.contains(pattern).all()
```

```python
burm.sample(10)
```

```python
# Combining the English and Burmese Dataframes (they are not Series) into a new dataframe, 
# via turning them into a  list (nuts, I know, but it's what worked) 
en_burm_trans = pd.DataFrame({'en':list(en.index),'burm':list(burm.index)})
en_burm_trans.sample(10)
```

```python
# The imported csv has '***' and '[Translation needed]' sometimes when a Translation is missing so fixing that
en_burm_trans.burm = en_burm_trans.burm.replace({'\*\*\*': np.nan, '***': np.nan, '[Translation needed]': np.nan})
```

```python
#writing the en_burm_trans df out to csv to check 
en_burm_trans.to_csv("en_burms_trans.csv", encoding="utf8")
```

```python
def write_dict_to_yaml(filename, dic_output):
    """
    Writes dictionary data to a yaml file in utf-8 encoding
    
    Parameters:
        filename (str): the filename to be written
        df (pd.Dataframe): df data to be written in the yaml file
    """
    with open(f'{filename}.yml', 'w', encoding="utf-8") as outfile:
        yaml.dump(dic_output, outfile, encoding='utf-8', allow_unicode=True, default_flow_style=False) #default_flow_style=False
        
def write_df_to_yaml(filename, df, mode='a'):
    """
    Writes dictionary data to a yaml file in utf-8 encoding
    
    Parameters:
        filename (str): the filename to be written
        dic_output (dict) the dictionary data to be written in the yaml file
    """
    import ipdb; ipdb.set_trace()
    dic_output = df.to_dict()
    #write yamls in append mode
    with open(f'{filename}.yml', mode, encoding="utf-8") as outfile:
        yaml.dump(dic_output, outfile, encoding='utf-8', allow_unicode=True, default_flow_style=False) #default_flow_style=True

```

```python
def get_all_translations():
    """
    Grabs all of the official English from the downloaded "en" translations yaml files 
    which will provide:
    1) the English text 
    2) the corresponding key. 
    
    The keys can be used to build the Burmese yaml later. 
    
    returns:
        dict of all data from yaml files
    """
    source_yaml_dir = "official_translations\\en"
    all_yamls = {}
    for file in os.listdir(source_yaml_dir):
        file_name = file[:-4]
        f = open(os.path.join(source_yaml_dir, file), 'r',  encoding="utf8")
        all_yamls[file_name] = {}
        for key,value in yaml.load(f).items():
            all_yamls[file_name][key] = value
    return all_yamls

all_yamls = get_all_translations()
```

```python
# Now we have all the translations from the yamls, we can make them into a dataframe
# This dataframe is needed for searching the values (translations) and picking the correct key
# Require structure is 
# --------------------------------------
# ¦ yamlfilename ¦ key_word ¦ sentence ¦
# ¦ -----------------------------------
all_en_translations = (pd.DataFrame(data=[(file_name, key_word, sentence)
                                       for file_name in all_yamls.keys()
                                       for key_word,sentence in all_yamls[file_name].items()],
                                 columns=['yaml_filename', 'key_word', 'sentence']))
```

```python
#changing the name of column 'sentence' to 'en' in the all_translations df
all_en_translations["en"]=all_en_translations["sentence"]
all_en_translations.drop("sentence", axis=1, inplace=True)
```

```python
all_en_translations.sample(20)
```

```python
# merging the dataframes on 'en'
merged_df = pd.merge(all_en_translations, en_burm_trans, on="en", how="outer").drop_duplicates().reset_index(drop=True)
merged_df.sample(20)
```

```python
#grabbing those with empty cells in the columns of interest
empty_burms = merged_df[pd.isnull(merged_df.burm)]
empty_key_words = merged_df[pd.isnull(merged_df.key_word)]
```

```python
#writing each of the dataframes containing empty cells to their own csvs
empty_burms.to_csv("empty_burms.csv", encoding="utf-8-sig", index=False)
empty_key_words.to_csv("empty_key_words.csv", encoding="utf-8-sig", index=False)
```

```python
#Now dropping those with nans, since we've dumped them to a csv
merged_df.dropna(axis=0, inplace=True)
```

```python
# empty results dictionary
res = {}
# split data into a dataframe for each yaml file
for yaml_file_name in merged_df.yaml_filename.unique():
    df = merged_df[merged_df['yaml_filename'] == yaml_file_name]
    df = df.drop(['en','yaml_filename'], axis=1)
    res[yaml_file_name] = df
    
```

```python
# Try to write each row to a dictionary, then to yamls
for yaml_filename in res.keys():
    print(yaml_filename)
#     import ipdb; ipdb.set_trace()
    d = res[yaml_filename].set_index("key_word").to_dict()['burm']
    write_dict_to_yaml(yaml_filename, dic_output=d)
```
